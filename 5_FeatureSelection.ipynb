### Imports

```python
import pandas as pd

from python_scripts.machine_learning import *
```

### Functions

```python
def call_feature_selection(fn, metric, zebra_level='0.1'):
    
    #Metadata
    metadata = pd.read_csv('processed_data/metadata/metadata_' + fn + '.tsv', sep='\t')
    
    #biom tables in df format
    df_wol2_genome = pd.read_csv('processed_data/pandas_df/' + fn + '_qiita15336_prep16181_pangenome_wol2_scrubbed_zebraFilter' + zebra_level + '.tsv' , sep='\t', index_col = 0)
   
    #Feature selection
    call_all_feature_selection([df_wol2_genome],
                            [fn + '_wol2_genome_pangenome'], 
                            metadata, metric, 'wol2')
    
    return()
```

```python
def call_machine_learning(fn, metric, zebra_level='0.1'):
    
    #Metadata
    metadata = pd.read_csv('processed_data/metadata/metadata_' + fn + '.tsv', sep='\t')
    
    #biom tables in df format
    df_wol2_genome = pd.read_csv('processed_data/pandas_df/' + fn + '_qiita15336_prep16181_pangenome_wol2_scrubbed_zebraFilter' + zebra_level + '.tsv' , sep='\t', index_col = 0)
   
    #Feature selection
    call_all_ML([df_wol2_genome],
                [fn + '_wol2_genome_pangenome'], 
                metadata, metric, 
                cross_validation_splits=3, gb_max_depth = 2, gb_n_estimators = 100, loops = 50)
    
```
